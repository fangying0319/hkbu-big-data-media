```python
import requests
import bs4
import csv
r=requests.get('https://movie.douban.com/top250')
```

```python
from bs4 import BeautifulSoup
mypage=BeautifulSoup(r.text)
```

```python
page_next=[]
url = 'https://movie.douban.com/top250'
for q in [-25,0,25,50,75,100,125,150,175,200,225]:
    page_number=str(q + 25)
    url_next=url+ "?start=" + page_number + '&filter='
    page_next.append(url_next)
```

```python
data = []
for page in page_next:
    l1 = requests.get(page).text
    mypage = BeautifulSoup(l1,'html.parser')
    mylis = mypage.find('ol', attrs={'class': 'grid_view'}).find_all('li')

    for li in mylis:
        myspans = li.find('div', attrs={'class': 'hd'}).find_all('span')
        cn_name = myspans[0].text
        en_name = myspans[1].text[3:]
        comment_number = li.find('div', attrs={'class': 'star'}).find_all('span')[3].text[:-3]
        rating = li.find('span', attrs={'class': 'rating_num'}).text
        director = li.find('p').text.strip()
#         idx = director.find('主演')
#         b,c = director[:idx],director[idx:]
       
        data.append([cn_name, en_name,  rating, comment_number , director])
```

```python
with open('data_new.csv','w') as f:
    writer = csv.writer(f)
    header = ['cn_name','en_name','rating' , 'comment_number' , 'director']
    writer.writerow(header)
    writer.writerows(data)
```
